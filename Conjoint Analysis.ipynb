```python
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
#Data
df = pd.read_csv('netflix_customer_survey.csv')
df.head()
```

#Prepare the Data

```python
#Isolate X and Y
y = df['selected']
#specifcy columns I don't want
x = df.drop(columns = ['selected', 'customerid'])
x.head(2)
```

```python
#Transform dummy variables
x_dummy = pd.get_dummies(x, columns = x.columns)
x_dummy.head()
```

#Logistic Regression

```python
#regression
model = sm.OLS(y, x_dummy, family = sm.families.Binomial()).fit()
model.summary()
```
